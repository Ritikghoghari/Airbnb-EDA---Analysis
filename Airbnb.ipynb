# Task 4 .Data Cleaning

```python
data.isnull().sum()

# droping all mising value
data.dropna(inplace=True)

data.isnull().sum()
```

```python
data.shape
```

```python
#dealing with duplication value
data.duplicated().sum()

# delete all duplicating value
data.drop_duplicates(inplace = True)
# data[data.duplicated()]

data.duplicated().sum()
```

```python
data[data.duplicated()]
```

```python
data.dtypes
```

```python
#type casting
# changing data types
data['id'] = data['id'].astype(object)

data['host_id'] = data['host_id'].astype(object)
data.dtypes
```

# Task 5 . Data Analysis

**Exploary data Analysis**


**Univariate Analysis**

```python
# indentifying outliers in price

df = data[data['price'] < 1500]

sns.boxplot(data=df , x='price')
```

```python
# Price Distribution

plt.figure(figsize=(8,5))
plt.title('Price Distribution')
plt.ylabel('Frequency')
sns.histplot(data=df , x='price' , bins=100)
plt.show()
```

```python
df.dtypes
```

```python
plt.figure(figsize=(6,3))
plt.title('availability_365 Distribution')
plt.ylabel('Frequency')
sns.histplot(data=df , x='availability_365')
plt.show()
```

```python
data.dtypes
```

```python
# average price per beds

df.groupby(by='neighbourhood_group')['price_per_bed'].mean()
```

```python
df.groupby(by='neighbourhood_group')['price'].mean()
```

# Feature Engineering 

```python
df['price_per_bed'] = df['price']/df['beds']
df.head()
```

# Bi Variable Analysis

**One variable dependency in another variable**

```python
df.columns
```

```python
# Price denpendency on neighbourhood
sns.barplot(data=df , x = 'neighbourhood_group' , y = 'price' , hue='room_type')
plt.title("Price Dependency on Neighbourhood")
plt.show()
```

```python
# Number of Reviews and price

plt.figure(figsize=(8,5))
plt.title("Locality and Review Dependency")
sns.scatterplot(data=df , x = 'number_of_reviews' , y ='price' , hue='neighbourhood_group')
```

```python
df.dtypes
```

```python
sns.pairplot(data=df , vars=['price' , 'minimum_nights','number_of_reviews', 'availability_365'], hue='room_type')
```

```python
# Cretae Gieography Charts
plt.figure(figsize=(10,7))
sns.scatterplot(data=df , x='longitude' , y ='latitude' , hue='room_type')
plt.title('Geographical Distribution of Airbnb Listing')
plt.show()
```
